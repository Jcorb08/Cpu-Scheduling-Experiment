```python
# Graphs
import pandas as pd

def calc_avg_Fcfs(test,i):
    return (FcfsDict[0][test][i+1]+FcfsDict[1][test][i+1]+FcfsDict[2][test][i+1])/3

def calc_avg_RR(test,i):
    return (RoundRobinDict[0][test][i+1]+RoundRobinDict[1][test][i+1]+RoundRobinDict[2][test][i+1])/3

def calc_avg_Ideal(test,i):
    return (IdealJobDict[0][test][i+1]+IdealJobDict[1][test][i+1]+IdealJobDict[2][test][i+1])/3

def calc_avg_F(test,i):
    return (FeedbackDict[0][test][i+1]+FeedbackDict[1][test][i+1]+FeedbackDict[2][test][i+1])/3

def calc_avg_Short(test,i):
    return (ShortJobDict[0][test][i+1]+ShortJobDict[1][test][i+1]+ShortJobDict[2][test][i+1])/3

def get_calc(test,i,scheduler):
    if scheduler == 0:
        return calc_avg_Fcfs(test,i)
    elif scheduler == 1:
        return calc_avg_RR(test,i)
    elif scheduler == 2:
        return calc_avg_Ideal(test,i)
    elif scheduler == 3:
        return calc_avg_F(test,i)
    else:
        return calc_avg_Short(test,i)

def get_data(processes):
    graph_data = {'Process ID': [i+1 for i in range(processes)]}
    if processes == 10:
        for i in range(5):
            graph_data['{}'.format(schedulers[i])] = [
            get_calc(0,j,i) for j in range(10)]    
    elif processes == 20:
        for i in range(5):
            graph_data['{}'.format(schedulers[i])] = [
            get_calc(1,j,i) for j in range(20)]
    elif processes == 30:
        for i in range(5):
            graph_data['{}'.format(schedulers[i])] = [
            get_calc(2,j,i) for j in range(30)]
    elif processes == 40:
        for i in range(5):
            graph_data['{}'.format(schedulers[i])] = [
            get_calc(3,j,i) for j in range(40)]
    else:
        for i in range(5):
            graph_data['{}'.format(schedulers[i])] = [
            get_calc(4,j,i) for j in range(50)]
    return graph_data

for processes in iLength:
    graph = pd.DataFrame(get_data(processes))
    graph.set_index('Process ID', inplace = True)
    graph_plot = graph.plot(title = 'Avg Response Time for {} Processes'.format(processes), figsize = (20,10))
    graph_plot.set_ylabel("Response Time")
```
