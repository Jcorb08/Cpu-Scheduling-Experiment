```python
# Graphs
import pandas as pd

def calc_avg_Fcfs(test,i):
    return (FcfsDict[0][test][i+1]+FcfsDict[1][test][i+1]+FcfsDict[2][test][i+1])/3

def calc_avg_Short(test,i):
    return (ShortJobDict[0][test][i+1]+ShortJobDict[1][test][i+1]+ShortJobDict[2][test][i+1])/3

def get_calc(test,i,scheduler):
    if scheduler == 0:
        return calc_avg_Fcfs(test,i)
    else:
        return calc_avg_Short(test,i)

def get_data(burst):
    graph_data = {'Process ID': [i+1 for i in range(30)]}
    if burst == 5:
        for i in range(2):
            graph_data['{}'.format(schedulers[i])] = [
            get_calc(0,j,i) for j in range(30)]    
    elif burst == 10:
        for i in range(2):
            graph_data['{}'.format(schedulers[i])] = [
            get_calc(1,j,i) for j in range(30)]
    elif burst == 15:
        for i in range(2):
            graph_data['{}'.format(schedulers[i])] = [
            get_calc(2,j,i) for j in range(30)]
    elif burst == 20:
        for i in range(2):
            graph_data['{}'.format(schedulers[i])] = [
            get_calc(3,j,i) for j in range(30)]
    else:
        for i in range(2):
            graph_data['{}'.format(schedulers[i])] = [
            get_calc(4,j,i) for j in range(30)]
    return graph_data

for burst in burstTime:
    graph = pd.DataFrame(get_data(burst))
    graph.set_index('Process ID', inplace = True)
    graph_plot = graph.plot(title = 'Avg Turnaround Time for a avg Burst Time of {}'.format(burst), figsize = (20,10))
    graph_plot.set_ylabel("Turnaround Time")
```
