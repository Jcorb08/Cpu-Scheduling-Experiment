```python
# Graphs
import pandas as pd

def calc_avg_RR(test,i):
    return (RoundRobinDict[0][test][i+1]+RoundRobinDict[1][test][i+1]+RoundRobinDict[2][test][i+1])/3

def calc_avg_F(test,i):
    return (FeedbackDict[0][test][i+1]+FeedbackDict[1][test][i+1]+FeedbackDict[2][test][i+1])/3

def get_data(scheduler):
    graph_data = {'Time Quantum': TimeQuantum}
    # round robin
    if scheduler == 0:
        for i in range(10):
            graph_data['Process Id {}'.format(i+1)] = [
            calc_avg_RR(0,i),
            calc_avg_RR(1,i),
            calc_avg_RR(2,i),
            calc_avg_RR(3,i),
            calc_avg_RR(4,i)
            ]    
    # feedback
    else:
        for i in range(10):
            graph_data['Process Id {}'.format(i+1)] = [
            calc_avg_F(0,i),
            calc_avg_F(1,i),
            calc_avg_F(2,i),
            calc_avg_F(3,i),
            calc_avg_F(4,i)
            ]
        
    return graph_data

schedulers = ['Round Robin', 'Feedback Round Robin']
for scheduler in range(2):
        graph = pd.DataFrame(get_data(scheduler))
        graph.set_index('Time Quantum', inplace = True)
        graph_plot = graph.plot(title = 'Avg Turnaround Time for {}'.format(schedulers[scheduler]), figsize = (20,10))
        graph_plot.set_ylabel("Turnaround Time")
```
